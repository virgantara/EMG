```python
base_path = 'DATARAW'

channels = ['GM','TA','VM']

features_names = ['VAR', 'RMS', 'IEMG', 'MAV', 'LOG', 'WL', 'ACC', 'DASDV', 'ZC', 'WAMP', 'MYOP', "FR", "MNP", "TP",
                      "MNF", "MDF", "PKF", "WENT"]

sampling_frequency = 1024
frame = 120
step = 60
x_gm, x_ta, x_vm = [],[],[]
x = []
# writer = pd.ExcelWriter('output_sampling'+str(sampling_frequency)+'_frame'+str(frame)+'_step'+str(step)+'.xlsx')
df = pd.DataFrame(columns=features_names)
for path in sorted(os.listdir(base_path)):
    subpaths = os.path.join(base_path,path)
    
    for subpath in os.listdir(subpaths):
        
        dir_path = os.path.join(base_path,path,subpath)
        
        for fname in os.listdir(dir_path):
            file_path = os.path.join(base_path,path,subpath,fname)

            tmp = []
            if ' GM ' in fname:
                emg_signal = []
                for line in open(file_path):
                    line = line.replace("\n","")
                    sinyal = line.split("\t")
                    emg_signal.append(sinyal[1])
                    tmp.append(np.array(['GM',sinyal[0],sinyal[1]]))
                
                signal = np.array(emg_signal).astype(np.float32)
                signal = signal.reshape((signal.size,))
                filtered_signal = signal
                emg_features, features_names = features_estimation(filtered_signal, 'GM',sampling_frequency, frame, step)
                emg_features = emg_features.T
                d = pd.DataFrame(emg_features)
                d.columns = features_names
                d['file'] = fname
                d['subject'] = path
                d['otot'] = 'GM'
                
                if ' Barefoot' in fname:
                    d['footwear'] = 'Barefoot'
                elif ' ShoeA' in fname:
                    d['footwear'] = 'ShoeA'
                elif ' ShoeB' in fname:
                    d['footwear'] = 'ShoeB'
                elif ' ShoeC' in fname:
                    d['footwear'] = 'ShoeC'
                
                df = df.append([d],ignore_index = True)
                


            if ' TA ' in fname:
                emg_signal = []
                for line in open(file_path):
                    line = line.replace("\n","")
                    sinyal = line.split("\t")
                    emg_signal.append(sinyal[1])
                    tmp.append(np.array(['TA',sinyal[0],sinyal[1]]))
                
                signal = np.array(emg_signal).astype(np.float64)
                signal = signal.reshape((signal.size,))
                filtered_signal = signal
                emg_features, features_names = features_estimation(filtered_signal, 'TA',sampling_frequency, frame, step)
                emg_features = emg_features.T
                d = pd.DataFrame(emg_features)
                d.columns = features_names
                d['file'] = fname
                d['subject'] = path
                d['otot'] = 'TA'
                if ' Barefoot' in fname:
                    d['footwear'] = 'Barefoot'
                elif ' ShoeA' in fname:
                    d['footwear'] = 'ShoeA'
                elif ' ShoeB' in fname:
                    d['footwear'] = 'ShoeB'
                elif ' ShoeC' in fname:
                    d['footwear'] = 'ShoeC'
                df = df.append([d],ignore_index = True)
            if ' VM ' in fname:
                emg_signal = []
                for line in open(file_path):
                    line = line.replace("\n","")
                    sinyal = line.split("\t")
                    emg_signal.append(sinyal[1])
                    tmp.append(np.array(['VM',sinyal[0],sinyal[1]]))
                
                signal = np.array(emg_signal).astype(np.float64)
                signal = signal.reshape((signal.size,))
                filtered_signal = signal
                emg_features, features_names = features_estimation(filtered_signal, 'VM',sampling_frequency, frame, step)
                emg_features = emg_features.T
                d = pd.DataFrame(emg_features)
                d.columns = features_names
                d['file'] = fname
                d['subject'] = path
                d['otot'] = 'VM'
                if ' Barefoot' in fname:
                    d['footwear'] = 'Barefoot'
                elif ' ShoeA' in fname:
                    d['footwear'] = 'ShoeA'
                elif ' ShoeB' in fname:
                    d['footwear'] = 'ShoeB'
                elif ' ShoeC' in fname:
                    d['footwear'] = 'ShoeC'
                df = df.append([d],ignore_index = True)
df.to_csv('output_sampling'+str(sampling_frequency)+'_frame'+str(frame)+'_step'+str(step)+'.csv')
writer.save()               
```
